2. DataFrame Fundamentals

A DataFrame is like a table with rows and columns.

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Priya"],
    "Age": [20, 21]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 4. Loading CSV Files
import pandas as pd

df = pd.read_csv("data.csv")
print(df)
```

```python
# 5. Exploring a DataFrame
print(df.head())      # First 5 rows
print(df.tail())      # Last 5 rows
print(df.info())      # Column information
print(df.describe())  # Statistics
```

```python
# 6. Indexing By Row & Column
print(df["Name"])      # Select column

print(df.loc[0])       # First row

print(df.loc[0, "Name"])  # Specific value
```

```python
# 7. Filtering
result = df[df["Age"] > 20]
print(result)
```

```python
# 8. Updating Data
df.loc[0, "Age"] = 25

print(df)
```

```python
# 9. Deleting Data

# Delete a column:

df.drop("Age", axis=1, inplace=True)

print(df)
```
