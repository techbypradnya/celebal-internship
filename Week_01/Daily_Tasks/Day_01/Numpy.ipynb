2. What is NumPy?

NumPy (Numerical Python) is a library for numerical computing using fast multidimensional arrays.

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

3. Why Use NumPy?

Faster than Python lists and uses less memory.

```python
import numpy as np

list1 = [1, 2, 3, 4]
array1 = np.array([1, 2, 3, 4])

print(array1 * 2)
```

4. Creating NumPy Arrays

```python
import numpy as np

a = np.array([1, 2, 3])

b = np.zeros(5)

c = np.ones(5)

d = np.arange(1, 10)

e = np.linspace(0, 1, 5)

print(a)
print(b)
print(c)
print(d)
print(e)
```

5. NumPy Data Types

```python
import numpy as np

arr = np.array([1, 2, 3], dtype=np.float64)

print(arr.dtype)
```

6. Multi-Dimensional Arrays

```python
#2D
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr)
```

```python
#3D
arr3d = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print(arr3d)
```

7. Array Attributes

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

8. Indexing & Slicing

```python
#Indexing
arr = np.array([10,20,30,40])

print(arr[0])
print(arr[-1])
```

```python
#Slicing
print(arr[1:3])
```
