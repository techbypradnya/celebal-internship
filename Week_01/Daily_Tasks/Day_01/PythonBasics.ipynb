Applications of Python

Data Types

```python
# Integer
age = 20

# Float
price = 99.99

# String
name = "Rahul"

# Boolean
is_student = True

# List
fruits = ["Apple", "Banana", "Mango"]

# Tuple
colors = ("Red", "Green", "Blue")

# Dictionary
student = {"name": "Rahul", "age": 20}

# Set
numbers = {1, 2, 3, 4}

# Display values and their data types
print(age, "-", type(age))
print(price, "-", type(price))
print(name, "-", type(name))
print(is_student, "-", type(is_student))
print(fruits, "-", type(fruits))
print(colors, "-", type(colors))
print(student, "-", type(student))
print(numbers, "-", type(numbers))
```

Arithmetic Operators

```python
a = 10
b = 5

print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division
```

String Methods

```python
text = "python"

print(text.upper())
print(text.lower())
print(text.capitalize())
```

Conditional Operators

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a < b)
print(a > b)
```

Chained Conditionals

```python
age = 20
student = True

if age >= 18 and student:
    print("Eligible")
```

If, Else, Elif Statements

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

For Loops

```python
for i in range(5):
    print(i)
```

While Loops

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

Comprehensions

```python
squares = [x*x for x in range(5)]

print(squares)
```

Functions

```python
def greet():
    print("Hello")

greet()
```

*args

```python
def add(*args):
    print(sum(args))

add(10, 20, 30)
```

**kwargs

```python
def info(**kwargs):
    print(kwargs)

info(name="Rahul", age=20)
```

Scope and Global Variables

```python
x = 10

def show():
    print(x)

show()
```

Exceptions

```python
try:
    num = 10 / 0
except:
    print("Error Occurred")
```

Lambda Functions

```python
square = lambda x: x * x

print(square(5))
```

Map Function

```python
numbers = [1, 2, 3]

result = list(map(lambda x: x * 2, numbers))

print(result)
```

Filter Function

```python
numbers = [1, 2, 3, 4, 5]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```
