**Your Geometric Explanation (4.3):**

> *Replace this text with your explanation of what eigenvalues/eigenvectors mean geometrically — what does it mean for a matrix to "stretch" or "squish" a vector?*

When you apply a matrix to a vector, it usually changes both its direction and length. But some special vectors only get scaled : their direction stays untouched. Those are eigenvectors.

The eigenvalue is simply the scaling factor:     
λ = 2 -> vector stretches to double    
λ = 0.5 -> vector shrinks to half    
λ = −1 -> vector flips but stays on the same line    

Think of it this way : if a matrix is a wind blowing through space, most vectors get swept off course. 
But eigenvectors are like arrows aligned with the wind : they only get pushed forward or pulled back, never sideways.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt =np.linalg.svd(X, full_matrices=False)

X_reconstructed =U @ np.diag(S) @ Vt
X_approx        = np.outer(U[:, 0] * S[0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

When we decompose a matrix A = U · S · Vᵀ, each part plays a different role:

- Vᵀ (shape 3×3) — Each row is a principal component direction. 
The first row points along the direction of maximum variance, the second captures the next most important direction, and so on. 
This is the key matrix in PCA.

- S (shape 3,) — Contains the singular values [2.376, 0.913, 0.813] which act as importance scores.
The larger the value, the more variance is captured along that direction. 
Squaring them gives the actual variance explained by each component.

- U (shape 4×3) — Shows how each data point projects onto the new principal component axes — essentially your data in the new transformed space.

**The PCA Link:** When data is centered, SVD and PCA are the same operation. The rows of **Vᵀ** are exactly the principal components — identical to the eigenvectors you'd get from decomposing the covariance matrix traditionally.

> **Bottom line:** **Vᵀ holds the directions of maximum variance**, **S tells you how significant each direction is**, and **U tells you where your data points land** in the new space.
